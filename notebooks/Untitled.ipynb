```python
tables = pd.read_html("https://www.gov.br/prf/pt-br/acesso-a-informacao/dados-abertos/dados-abertos-da-prf", extract_links="body")
```

```python
table_for_download = tables[1].copy()
table_for_download = table_for_download.iloc[1:-1]
```

```python
table_for_download.columns = ["DESCRIÇÃO", "URL"]
```

```python
URL_REPLACE = "https://drive.usercontent.google.com/u/0/uc?id=ID_FILE&export=download"

table_for_download.loc[:, "URL"] = table_for_download["URL"].apply(lambda x: x[1].split("/")[-3])
table_for_download.loc[:, "URL"] = table_for_download["URL"].apply(lambda x: URL_REPLACE.replace("ID_FILE", x))
```

```python
table_for_download.loc[:, "DESCRIÇÃO"] = table_for_download["DESCRIÇÃO"].apply(lambda x: x[0])
```

```python
table_for_download.loc[:, 'DESCRIÇÃO'] = table_for_download['DESCRIÇÃO'].str.replace('Documento CSV de Acidentes ', '', regex=False)
```

```python
table_for_download['ANO'] = table_for_download['DESCRIÇÃO'].str.extract(r'(\d{4})')
```

```python
def filtrar_links(df_g):
    if df_g['DESCRIÇÃO'].str.contains('Todas as causas e', na=False).any():
        return df_g[df_g['DESCRIÇÃO'].str.contains('Todas as causas e', na=False)]
    else:
        return df_g

table_for_download = table_for_download.groupby('ANO').apply(filtrar_links, include_groups=True).reset_index(drop=True)
```

```python
def get_name_file(col):
    ano_number = col[:4]
    
    if "Todas as causas e tipos de acidentes" in col:
        return f"{col[:4]}_geral"
    elif "Agrupados por pessoa" in col:
        return f"{col[:4]}_agg_pessoa"
    else:
        return f"{col[:4]}_agg_ocorrencia"
    
table_for_download.loc[:, "FILE_NAME"] = table_for_download['DESCRIÇÃO'].apply(get_name_file)
```
