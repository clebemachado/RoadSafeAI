Verificando colunas de todos os datasets

```python
def analise_colunas(ano_inicio, ano_fim):
  colunas_por_ano = {}

  for ano in range(ano_inicio,ano_fim+1):
    path = f'D:/projetos/roadSafeAi/RoadSafeAI/data/raw/datatran{ano}.csv'
    
    df = pd.read_csv(path, encoding='latin1', sep=";", nrows=0)
    colunas_por_ano[ano] = set(df.columns)
    print(f"{ano}: {len(df.columns)} colunas")
    
  if(colunas_por_ano):
    ano_ref = min(colunas_por_ano.keys())
    coluna_ref = colunas_por_ano[ano_ref]
    
    print(f"\nComparando com o ano de referência ({ano_ref}):")
    
    for ano, colunas in colunas_por_ano.items():
      if ano != ano_ref:
        novas_colunas = colunas - coluna_ref
        colunas_faltantes = coluna_ref - colunas
        
        if novas_colunas or colunas_faltantes:
          print(f"\nDiferenças encontradas no ano {ano}:")
          if novas_colunas:
              print(f"Novas colunas: {novas_colunas}")
          if colunas_faltantes:
              print(f"Colunas ausentes: {colunas_faltantes}")
        else:
          print(f"\nAno {ano}: Mesmas colunas que {ano_ref}")
      
```

```python
analise_colunas(2007,2024)
```

É possivel notar uma alteração nas colunas entre os anos 2015, 2016 e 2017

```python
analise_colunas(2015,2017)
```

No ano de 2016, a coluna ano não foi declarada.

A partir do ano de 2017, as colunas referentes a região do acidente, como latitude e longitude começaram a ser registradas.

É possivel diante dessas diferenças, criar dois datasets, um com as 26 colunas que existem de 2007 a 2024 (adicionando a coluna ano no dataset de 2016) e criar uma coluna com dados referente a região, com os dataset de 2017 a 2024.
